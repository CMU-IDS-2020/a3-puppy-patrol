```python
import pandas as pd
import datetime
import numpy as np
import altair as alt
import math
```

```python
df = pd.read_csv("rats.csv")
```

```python
df
```

```python
fmt = lambda d : datetime.datetime.strptime(d, "%m-%d-%Y %H:%M:%S %p")
```

```python
def cmp_days(data):
    created = fmt(data['created_date'])
    closed = fmt(data['closed_date'])
    dlt = closed - created
    return dlt.days
```

```python
df['days_open'] = df.apply(lambda d: cmp_days(d) if type(d['created_date']) is str and type(d['closed_date']) is str else "NaN", axis=1)
```
